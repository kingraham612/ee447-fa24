# 1. matrix exponential

(a.) Show that matrix multiplication does not generally *commute*, i.e. find $A,B\in\mathbb{R}^{n\times n}$ for which $A\,B \neq B\,A$.  (*Hint:  you should be able to find an example when $n = 2$.*)

**Solution:**

 Counterexample:     

 Let $$A = \left[\begin{matrix} 1 & 2 \\ -4 & 3 \end{matrix}\right]  and B = \left[\begin{matrix} 2 & 5 \\ 9 & -8 \end{matrix}\right]$$   

$AB =  \left[\begin{matrix} 20 & -11 \\ 19 & -44 \end{matrix}\right] $   
$BA =  \left[\begin{matrix} -18 & 19 \\ 41 & -6 \end{matrix}\right] $   

As we can see, $$AB \neq BA$$


 For square matrices, here is a general solution for n = 2.    

 $$A = \left[\begin{matrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{matrix}\right]  and B = \left[\begin{matrix} b_{11} & b_{12} \\ b_{21} & b_{22} \end{matrix}\right]$$

 $$AB =  \left[\begin{matrix} a_{11}b_{11}+a_{12}b_{21} & a_{12}b_{22}+a_{11}b_{12} \\ a_{21}b_{11}+a_{22}b_{21} & a_{21}b_{12}+a_{22}b_{22} \end{matrix}\right]$$

 $$BA =  \left[\begin{matrix} b_{11}a_{11}+b_{12}a_{21} & b_{11}a_{12}+b_{12}a_{22} \\ b_{21}a_{11}+b_{22}a_{21} & b_{12}a_{12}+b_{22}a_{22} \end{matrix}\right]$$

The result of $AB \neq BA$ until the value of products equal each other. Thus, matrix multiplication is not generally commutative.

**Now recall the definition of the *matrix exponential*:**

$$ \forall A \in \mathbb{R}^{n\times n} : e^A = \sum_{k=0}^\infty \frac{1}{k!} A^k. $$

**In the remainder of this problem, you'll use this definition directly to establish useful facts.**

(b.) Show that every square matrix commutes with its matrix exponential, i.e. $A\, e^A = e^AA$ for all $A\in\mathbb{R}^{n\times n}$.

**Solution:**

Starting from the L.H.S:   
$A\, e^A = A\, \sum_{k=0}^\infty \frac{1}{k!} A^k$

For square matrices, the matrix exponential can be re-written as:    
$A\, e^A  =  A(I + A+\frac{A^2}{2!}+ \frac{A^3}{3!}...) $    
$\Leftrightarrow  A + AA+\frac{AA^2}{2!}+ \frac{AA^3}{3!}...$  
$\Leftrightarrow  A + A^2+\frac{A^3}{2!}+ \frac{A^4}{3!}...$     

By factoring out A to the right,    
$\Leftrightarrow  A\, e^A = (I + A+\frac{A^2}{2!}+ \frac{A^3}{3!}...)A $   
$\Leftrightarrow A\, e^A =   e^AA$   

Therefore, square matrices commute with their matrix exponentials.

(c.) Given an invertible matrix $T\in\mathbb{R}^{n\times n}$, show that $e^{T A T^{-1}} = T e^A T^{-1}$.

**Solution:**

$e^{T A T^{-1}} = T e^A T^{-1}$   
$e^A = \sum_{k=0}^\infty \frac{1}{k!} A^k$     
$\Leftrightarrow e^{T A T^{-1}} = I + T A T^{-1}+\frac{(T A T^{-1})^2}{2!}+ \frac{(T A T^{-1})^3}{3!}...$     

By examining the terms in the infinite sum:    

$(T A T^{-1})^2 = T A T^{-1} T A T^{-1}$
$\Leftrightarrow (T A T^{-1})^2 = TAIAT^{-1}$ (Since $TT^{-1} = I$)    
$\Leftrightarrow (T A T^{-1})^2 = TA^2T^{-1}$

Similarly, for each of the product terms, we get $(TAT^{-1})^n = TA^nT^{-1}$.

Therefore,     
$\Leftrightarrow e^{T A T^{-1}} = I + T A T^{-1}+\frac{TA^2T^{-1}}{2!}+ \frac{TA^3T^{-1}}{3!}...$
$\Leftrightarrow e^{T A T^{-1}} = TIT^{-1} + TAT^{-1} +\frac{TA^2T^{-1}}{2!}+ \frac{TA^3T^{-1}}{3!}...$
$\Leftrightarrow e^{T A T^{-1}} = T(I + A+\frac{A^2}{2!}+ \frac{A^3}{3!}...)T^{-1}$
$\Leftrightarrow e^{T A T^{-1}T} = Te^AT^{-1}$

(d.) If $\lambda\in\mathbb{C}$ is an eigenvalue of $A$ with eigenvector $v\in\mathbb{C}^{n}$, show that $e^{\lambda t}$ is an eigenvalue of $e^{A t}$.

**Solution:**

To prove $e^{\lambda}$ is an eigenvalue of $e^A$ i.e, $e^Av = e^{\lambda}v$     
​
For, $Av = \lambda v$, $A^kv = \lambda^kv$
​
*Proof by induction:*   
    
$A^0 v=Iv=v=1*v=\lambda^0 v$    
On multiplying with $A$ on both sides:     
$Av = Av \Leftrightarrow Av = \lambda v$    
Upon multiplying with A for the above equation,    
$AAv = A \lambda v$     
$ \Leftrightarrow Av = \lambda (Av)$     
$\Leftrightarrow Av = \lambda (\lambda v)$
$A^2v = \lambda^2 v$
​
Taking a more general case:     
​
$A^{k-1}v = \lambda^{k-1} v$     
On multiplying with A to both sides,    
$ AA^{k-1}v=A (\lambda^{k-1}v)$     
$\Leftrightarrow A (A^{k-1} v)= \lambda^{k-1} (Av)$
​
$\Leftrightarrow A^k v = \lambda^{k}v$
​
​
Therefore, rewriting the formula with the property proven above,     
$ \sum_{k=0}^\infty \frac{1}{k!}A^k v =  \sum_{k=0}^\infty \frac{1}{k!} \lambda^k v$
​
Therefore,     
$e^Av=\sum_{k=0}^\infty \frac{1}{k!}A^k v$     
$ \Leftrightarrow \sum_{k=0}^\infty \frac{1}{k!} \lambda^k v$     
$ \Leftrightarrow e^{\lambda}v$

Evaluating the nonlinear equations using the parameter values yields: 
$$\left[\begin{matrix} \dot{q} \\ \frac{C}{M}\left(\frac{i}{q}\right)^2 - g \\ \frac{1}{L} \left ( -R i + 2 C \frac{i \dot{q}}{q^2} + u \right ) \end{matrix}\right]

The full Jacobian derivative of $f$ with respect to $x$ is
$$ A = \partial_x f = 
\left[\begin{matrix}

\end{bmatrix}
= \left[\begin{matrix} 0 & 1 & 0 \\ -0.016i^2q^{-3} & 0 & 0.016iq^{-2} \\ -4\times 10^{-4} i\dot{q}q^{-3} & 2\times 10^{-4}iq^{-2} & 2\times 10^{-4}\dot{q}q^{-2} - 30 \end{matrix}\right]$$

The Jacobian derivative of $f$ with respect to $u$ is
$$ B = \partial_u f = \left[\begin{matrix} \partial_{u} \dot{q} \\ \partial_{u} \ddot{q} \\ \partial_{u} \dot{i} \end{matrix}\right]
= \left[\begin{matrix} 0 \\ 0 \\ 1 \end{matrix}\right]$$

The Jacobian A matrix evaluated at $x_{eq} = (q_{eq},\dot{q}_{eq}, i_{eq})$  is:

```python
# import modules first
import numpy as np
import pylab as plt

q = 0.05
dq = 0
i = 1.72

A = np.array([[0, 1, 0],
              [-0.016*(i**2)*(q**(-3)), 0, 0.016*i*(q**(-2))],
              [-4*10**(-4)*i*dq*q**(-3), 2*10**(-4)*i*q**(-2), 2*10**(-4)*dq*q**(-2)-30]])
print('A')
print(A)
eigenvalue, _ = np.linalg.eig(A)
print('eigenvalue of A')
print(eigenvalue)
```

**Solution:** \
In order to get the system into the form $\delta \dot{x} = (A-BK) \delta x$ we need to write the control input $\delta u$ as a function of $\delta x$, not just $\delta q$. But we know that $\delta q$ is just the first term of $\delta x$, because $\delta x = [ \delta q,  \delta \dot{q}, \delta i ]^T$. Therefore, we can rewrite: 

$$ \delta u = -k_P \delta q = \begin{bmatrix}-k_P, 0, 0 \end{bmatrix} \begin{bmatrix} \delta q \\ \delta \dot{q} \\ \delta i \end{bmatrix}$$

Therefore $\delta \dot{x} = (A-BK) \delta x$ is:

$$ \delta \dot{x} = \left( A - B \begin{bmatrix}k_P, 0, 0 \end{bmatrix} \right) \begin{bmatrix} \delta q \\ \delta \dot{q} \\ \delta i \end{bmatrix} $$

Notice that $A \in \mathbb{R}^{3\times 3}, B \in \mathbb{R}^{3\times 1}$, and $K \in \mathbb{R}^{1\times 3}$, with $\delta x \in \mathbb{R}^{3}$ makes the dimensions work out.

(d.) Create a root locus plot for the closed-loop system as parameter $k_P$ varies.

```python
import numpy as np
import pylab as plt

q = 0.05
dq = 0
i = 1.72

#Choosing values of k for annotation

kps = np.linspace(-200,2000,70)

lambda1 = np.array([])
lambda2 = np.array([])
lambda3 = np.array([])
for kp in kps:
  A = np.array([[0, 1, 0],
              [-0.016*(i**2)*(q**(-3)), 0, 0.016*i*(q**(-2))],
              [-4*10**(-4)*i*dq*q**(-3)-kp, 2*10**(-4)*i*q**(-2), 2*10**(-4)*dq*q**(-2)-30]])

  eigenvalue, _ = np.linalg.eig(A)

  lambda1 = np.append(lambda1, eigenvalue[0])
  lambda2 = np.append(lambda2, eigenvalue[1])
  lambda3 = np.append(lambda3, eigenvalue[2])

print(np.shape(lambda1))
# confirm the non-real eigenvalues are complex-conjugates
assert np.allclose(lambda1[np.abs(lambda1.imag) > 0.],
                   np.conj(lambda2[np.abs(lambda1.imag) > 0]))

plt.figure(figsize=(8,8))

#Axis labels
plt.title('locus of eigenvalues as feedback parameter $k$ varies')
plt.xlabel(r'real part of eigenvalue ($\operatorname{Re}\ \lambda$)')
plt.ylabel(r'imaginary part of eigenvalue ($\operatorname{Im}\ \lambda$)')

#Plotting values
plt.scatter(lambda1.real,lambda1.imag,color='magenta')
plt.scatter(lambda2.real,lambda2.imag,color = 'aquamarine')
plt.scatter(lambda3.real,lambda3.imag,color = 'g')
# we'll only annotate every 25 of the computed eigenvalues
for k,lam1,lam2 in list(zip(kps,lambda1,lambda2))[::25]:
  plt.annotate(r'$k=%0.1f$'%k,xy=(lam1.real,lam1.imag))
  plt.annotate(r'$k=%0.1f$'%k,xy=(lam2.real,lam2.imag))
#x- and y- limits and drawing axes
ylim = plt.ylim()
plt.plot(np.zeros(2),ylim,'k--')
xlim = plt.xlim()
plt.plot(xlim,np.zeros(2),'k--')
plt.xlim(xlim); plt.ylim(ylim);
```

Consider the following model of a robot arm that consists of a rigid rod of length $\ell$ attached to a point mass $m$ at one end and a rotational joint at the other end,

$$ \ m \ell^2 \ddot{\theta} = m g \ell \sin\theta - \gamma\dot{\theta} + \tau, $$

where
$\gamma$ is a coefficient of friction for the rotational joint
and
$\tau$ is a torque applied by a motor attached to the rotational joint.

With state $x = (\theta,\dot{\theta})$ and input $u = \tau$, linearizing the dynamics around the vertical equilibrium $x_e = (\theta_e,\dot{\theta}_e) = (0,0)$, $u_e = \tau_e = 0$ yields the linear control system

$$ \dot{\delta x} = A \delta x + B \delta u $$

where $x \approx x_e + \delta x$, $u \approx u_e + \delta u$,

$$
A = \left[\begin{matrix} 0 & 1 \\ \frac{g}{\ell} & -\frac{\gamma}{m\ell^2} \end{matrix}\right], \
B = \left[\begin{matrix} 0 \\ \frac{1}{m\ell^2} \end{matrix}\right].
$$



a. Design a linear state-feedback law $\delta u = - K \delta x$ that gives a closed-loop characteristic polynomial $s^2 + 2\zeta\omega s + \omega^2$ that corresponds to a second-order system that is ***overdamped*** ($\zeta = 3/2$) and has ***natural frequency*** $\omega = 2$Hz.
**Solution:**

We know from class that we can obtain the characteristic polynomial of a closed-loop system by computing $a(s)=det(sI-(A-BK))$.

First compute $BK$ where $K=\left[\begin{matrix} k_1 & k_2 \end{matrix}\right]$

$$BK = \left[\begin{matrix} 0 \\ \frac{1}{m\ell^2} \end{matrix}\right]\left[\begin{matrix} k_1 & k_2 \end{matrix}\right] = \left[\begin{matrix} 0 & 0 \\ \frac{k_1}{m\ell^2} & \frac{k_2}{m\ell^2} \end{matrix}\right]$$

$$a(s)=det(sI-(A-BK))=det(\left[\begin{matrix} s & 0 \\ 0 & s\end{matrix}\right] - \left[\begin{matrix} 0 & 1 \\ \frac{g}{\ell}-\frac{k_1}{m\ell^2} & -\frac{\gamma+k_2}{m\ell^2} \end{matrix}\right]) = det(\left[\begin{matrix} s & -1 \\ -\frac{g}{\ell}+\frac{k_1}{m\ell^2} & s+\frac{\gamma+k_2}{m\ell^2} \end{matrix}\right])$$

We know that $det(\left[\begin{matrix} a & b \\ c & d\end{matrix}\right]) = ad-bc$ so this determinant is equal to:

$$a(s) = s^2+\frac{\gamma+k_2}{m\ell^2}s-(\frac{g}{\ell}-\frac{k_1}{m\ell^2})$$

Compare against $s^2 + 2\zeta\omega s + \omega^2$ to obtain that

$\omega^2=-(\frac{g}{\ell}-\frac{k_1}{m\ell^2}) \Leftrightarrow \fbox{$k_1=m\ell g + \omega^2 m\ell^2 = m\ell g + 4 m\ell^2$}$

and $\frac{\gamma+k_2}{m\ell^2}=2\zeta\omega \Leftrightarrow \fbox{$k_2=2\zeta\omega m \ell^2-\gamma=6m\ell^2-\gamma$}$

c. Implement a simulation of the closed-loop system using parameter values from (b.) to verify that the closed-loop dynamics are as expected.



```python
# build ODE solver
def numerical_simulation(f,t,x,t0=0.,dt=1e-4,ut=None,ux=None,utx=None,return_u=False):
  """
  simulate x' = f(x,u)

  input:
    f : R x X x U --> X - vector field
      X - state space (must be vector space)
      U - control input set
    t - scalar - final simulation time
    x - initial condition; element of X

    (optional:)
    t0 - scalar - initial simulation time
    dt - scalar - stepsize parameter
    return_u - bool - whether to return u_

    (only one of:)
    ut : R --> U
    ux : X --> U
    utx : R x X --> U

  output:
    t_ - N array - time trajectory
    x_ - N x X array - state trajectory
    (if return_u:)
    u_ - N x U array - state trajectory
  """
  t_,x_,u_ = [t0],[x],[]

  inputs = sum([1 if u is not None else 0 for u in [ut,ux,utx]])
  assert inputs <= 1, "more than one of ut,ux,utx defined"

  if inputs == 0:
    assert not return_u, "no input supplied"
  else:
    if ut is not None:
      u = lambda t,x : ut(t)
    elif ux is not None:
      u = lambda t,x : ux(x)
    elif utx is not None:
      u = lambda t,x : utx(t,x)

  while t_[-1]+dt < t:
    if inputs == 0:
      _t,_x = t_[-1],x_[-1]
      dx = f(t_[-1],x_[-1]) * dt
    else:
      _t,_x,_u = t_[-1],x_[-1],u(t_[-1],x_[-1])
      dx = f(_t,_x,_u) * dt
      u_.append( _u )

    x_.append( _x + dx )
    t_.append( _t + dt )

  if return_u:
    return np.asarray(t_),np.asarray(x_),np.asarray(u_)
  else:
    return np.asarray(t_),np.asarray(x_)
```

e. Design a linear observer

$$ \dot{\delta\widehat{x}} = A \delta\widehat{x} + B u + L(\delta y - \delta\widehat{y}),\ \delta\widehat{y} = C \delta\widehat{x} + D u $$

so that the error dynamics $\dot{e} = (A - LC) e$ has characteristic polynomial $s^2 + 2\xi\sigma s + \sigma^2$ that corresponds to a second-order system that is ***underdamped*** ($\xi = 1/2$) and has ***natural frequency*** $\sigma = 10$Hz

**Solution:**


To determine the characteristic polynomial of the error dynamics, let's compute $a(s)=det(sI-(A-LC))$.

First compute $LC$ where $L=\left[\begin{matrix} l_1 \\ l_2 \end{matrix}\right]$

$$LC = \left[\begin{matrix} l_1 \\ l_2 \end{matrix}\right]\left[\begin{matrix} \ell & 0 \end{matrix}\right] = \left[\begin{matrix} \ell l_1 & 0 \\ \ell l_2 & 0 \end{matrix}\right]$$

$$a(s)=det(sI-(A-LC))=
det\left( \left[\begin{matrix} s & 0 \\ 0 & s\end{matrix}\right]- 
\left( \left[\begin{matrix} 0 & 1 \\ \frac{g}{\ell} & -\frac{\gamma}{m\ell^2} \end{matrix}\right] -
\left[\begin{matrix} \ell l_1 & 0 \\ \ell l_2 & 0 \end{matrix}\right] \right) \right) $$
$$ = det \left( \left[\begin{matrix} s & 0 \\ 0 & s\end{matrix}\right] -
\left[\begin{matrix} \ell l_1 & 1 \\ \frac{g}{l} - \ell l_2 & - \frac{\gamma}{m\ell^2} \end{matrix}\right] \right)
= det(\left[\begin{matrix} s + \ell l_1 & -1 \\ -\frac{g}{\ell} + \ell l_2 & s + \frac{\gamma}{m\ell^2} \end{matrix}\right])$$

We know that $det\left( \left[\begin{matrix} a & b \\ c & d\end{matrix}\right] \right) = ad-bc$ so this determinant is equal to:

$$a(s) = (s + \ell l_1)(s + \frac{\gamma}{m \ell^2}) - (-1)(\frac{-g}{\ell} + \ell l_2) \\
= s^2 + (\frac{\gamma}{m\ell^2} + \ell l_1) s +
\frac{l_1 \gamma}{m\ell} - \frac{g}{\ell} + \ell l_2
$$

Matching coefficients of $a(s)$ with $s^2 + 2\xi\sigma s + \sigma^2$, and using $\xi = 1/2, \sigma=10$ we get:

<!-- $$\frac{\gamma}{m\ell^2} - \ell l_1 = 2\xi\sigma
\Leftrightarrow
l_1 = \frac{\gamma}{m \ell^2} - \frac{10}{\ell}$$
 -->
<!-- $$- l_1 \cdot \frac{\gamma}{m\ell} - \frac{g}{\ell} - \ell l_2 = \sigma^2
\Leftrightarrow
l_2 = \frac{1}{\ell}(-(\frac{\gamma}{m \ell^2})^2 + \frac{10\gamma}{m \ell^2} - \frac{g}{\ell} - 100)$$ -->


$2\xi\sigma = (\ell l_1 + \frac{\gamma}{m\ell^2}) \Leftrightarrow \fbox{$l_1 = ( \frac{10}{\ell} - \frac{\gamma}{m\ell^3})$}$

$\sigma^2 = -\frac{\gamma l_1}{m\ell} + \frac{g}{\ell} + \ell l_2$
$\Leftrightarrow
\fbox{$l_2 = \frac{100}{\ell} + \frac{\gamma}{m \ell^2} (\frac{-10}{\ell} + \frac{\gamma}{m\ell^3}) + \frac{g}{\ell^2}$}$

```python
# define parameters
m = 1 # kg
ell = 1 # m
g = 9.81 # m/sec/sec
gamma = 1

xi = 1/2
sigma = 10 # Hz

A = np.asarray([[0,1],[g/ell, -gamma/(m*ell**2)]])
B = np.asarray([[0],[1/(m*ell**2)]])
C = np.asarray([[ell, 0]])
D = np.asarray([[0]])

# roots = [-5+8.66*1j, -5-8.66*1j] #Enter value
sqrt = np.lib.scimath.sqrt # for complex roots
s1 = (-(2*xi*sigma) + sqrt((2*xi*sigma)**2 - 4*1*sigma**2))/2
s2 = (-(2*xi*sigma) - sqrt((2*xi*sigma)**2 - 4*1*sigma**2))/2
roots = [s1,s2]
roots

L = place(A.T,C.T,roots).T
print('L found using ctrl.place package:')
print(L)
# check to see if solution matches one found by deriving in a
# l1 = (10/ell) - (gamma / (m * ell**3))# Need to determine the value of l1
# l2 = ((100/ell) -(gamma * l1)/(m * ell**3) + (g/ell**2) ) # Need to determine the value of l2

# l1 = -10/ell + gamma/(m*ell**3)
# l2 = - sigma**2/ell + 10*gamma/(m*ell**3) - gamma**2/((m**2)*(ell**5)) - g/ell**2

l1 = 10/ell - gamma/(m*ell**3)
l2 = sigma**2/ell - 10*gamma/(m*ell**3) + gamma**2/((m**2)*(ell**5)) + g/ell**2

print('L found from e:')
print(l1,l2)
```
